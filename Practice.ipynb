#scaling of data
transforming your data so that it fits within a specific scale, like 0-100 or 0-1.
transforming your data so that it fits within a specific scale, like 0-100 or 0-1.
Scaling convert the data using the formula = (value — min value) / (max value — min value)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
original_data = pd.DataFrame(data_set['Estimated Cost'])
scale_data = pd.DataFrame(scaler.fit_transform(original_data))
```

```python
original_data
```

```python
scale_data
```

#Normalizing of data
Scaling just changes the range of your data. Normalization is a more radical transformation. The point of normalization is to change your observations so that they can be described as a normal distribution.
formula = (value — mean) / standard deviation

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
original_data = pd.DataFrame(data_set['Estimated Cost'])
scale_data = pd.DataFrame(scaler.fit_transform(original_data))
```

```python
original_data
```

```python
scale_data
```
